# L7d: Sampling and Training a Restricted Boltzmann Machine

In this lab, we will explore the feedforward and feedback passes of a restricted Boltzmann machine (RBM) and train it using the contrastive divergence algorithm on handwritten digit images.

__Why are we looking at this?__ The [restricted Boltzmann machine](https://en.wikipedia.org/wiki/Restricted_Boltzmann_machine) overcomes the computational limitations of the general Boltzmann machine by restricting connections to form a bipartite graph between visible and hidden layers. This restriction enables efficient training using the contrastive divergence algorithm, making RBMs practical for real-world applications such as dimensionality reduction and feature learning.

> __Learning Objectives__
>
> By the end of this lab, you should be able to:
> * __Implement RBM sampling:__ Execute feedforward and feedback passes through an RBM to generate visible and hidden state configurations from input data. Understand how block Gibbs sampling alternately updates hidden and visible layers to explore the model's learned probability distribution.
> * __Train using contrastive divergence:__ Apply the CD algorithm to learn weights $\mathbf{W}$ and biases $\mathbf{a}$, $\mathbf{b}$ that capture the distribution of training images. Approximate the gradient of the log-likelihood by comparing data-driven activations to model reconstructions without computing the intractable partition function.
> * __Evaluate reconstruction quality:__ Compare the RBM's reconstructed outputs before and after training to assess learning progress. Observe how training improves the network's ability to reconstruct corrupted or noisy input patterns from the handwritten digit dataset.

### Implementation
The notebook uses functions for Boltzmann machine operations:

> `decode(s; number_of_rows, number_of_columns)`: Converts a flattened state vector $\mathbf{s}$ back to an $n \times n$ image matrix. This local function (defined below) reshapes the vector, transposes it to match row-major ordering, and replaces $-1$ values with $0$ for grayscale display.

```julia
function decode(s::Vector{Int64}; number_of_rows::Int=28, number_of_columns::Int = 28)::Array{Int64,2}

    # reconstructed image -
    X = reshape(s, number_of_rows, number_of_columns) |> X -> transpose(X) |> Matrix;
    XŐā = replace(X, -1 => 0);
    
    # return -
    return XŐā
end
```

## Task 1: Load and Explore the Data
In this task, we'll load [the USPS handwritten digits dataset](https://en.wikipedia.org/wiki/MNIST_database#USPS_database) that our restricted Boltzmann machine will model. These are `16` $\times$ `16` grayscale images of handwritten numbers.

__Load the data__. Let's load [the USPS dataset by calling the `MyUSPSHandwrittenDigitImageDataset()` method](https://varnerlab.github.io/VLDataScienceMachineLearningPackage.jl/dev/data/#VLDataScienceMachineLearningPackage.MyUSPSHandwrittenDigitImageDataset), transform the data to the `{-1,1}` scale, and then return this scaled data to the caller in the `X::Array{Int,2}` array, where the images are encoded in the columns of the array.

> __Data__
> 
> * __What is in the `X` array?__ Each column of the `X::Array{Int,2}` array holds one of the `7291` handwritten images. Each row is a pixel value; a total of `256` pixels per image. We also return the `raw_data_dictionary::Dict{Int64, Vector{Float64}}` dictionary that holds the raw (untransformed) data vectors for each image.
> * __What's in the labels array?__ The `labels::Vector{Int}` array holds the actual digit (0-9) corresponding to each image in the `X` array.

We save the data in the `X::Array{Int,2}` array for later use in training the RBM. We also save the `labels` for later use in visualizing the original images.

```julia
X, labels = let

    # load the dataset
    results = MyUSPSHandwrittenDigitImageDataset(); 
    records = results.records;
    labels = results.labels;
    number_of_records = length(labels);

    # get the number of records -
    X = Array{Int,2}(undef, number_of_pixels, number_of_records);

    for i ‚ąą 1:number_of_records
        record = records[i];
        for k ‚ąą 1:number_of_pixels
            if (record[k] ‚Č§ 0.0)
                X[k,i] = -1;
            else
                X[k,i] = 1;
            end
        end
    end

    X[:,1:number_of_images_to_explore], labels[1:number_of_images_to_explore]
end;
```

__Probability distribution of the data__. In the code below, we want to be able to sample the data, for example, generate mini-batches of data to estimate the model parameters. However, we cannot compute the partition function even though our data vectors are relatively small, i.e., `16^2 = 256` pixels. 

> __Approximating the data distribution__
>
> Let's compute the probability of observing a pattern in the _data that we do have_, and use this to approximate $P(\mathbf{x})_{\text{data}}$. We compare each pattern to every other by computing a _dissimilarity score_. If the dissimilarity is below a threshold, the patterns are assumed to be similar. We compute an approximate probability of seeing _similar_ patterns [using the `softmax(...)` method](src/Compute.jl) and use this to initialize [a `Categorical` distribution exported by the `Distributions.jl` package](https://juliastats.org/Distributions.jl/latest/univariate/#Distributions.Categorical).

We save the categorical distribution in the `data_distribution::Categorical{Float64, Vector{Float64}}` variable.


```julia
(data_distribution,p) = let

    # initialize -
    number_of_samples = size(X, 2); # how many turns do we have
    őł = Array{Float64,1}(undef, number_of_samples);

    for i ‚ąą 1:number_of_samples
        sŠĶĘ = X[:,i]; # get the ith pixel pattern

        counter = 0;
        for j ‚ąą 1:number_of_samples
            s‚Īľ = X[:,j]; # get the jth pixel pattern

            # compute the difference -
            d = (1/(4*number_of_pixels))*sum((sŠĶĘ .- s‚Īľ).^2);            
            if (d ‚Č§ 0.20) # cutoff
                counter += 1;
            end
        end
        őł[i] = counter/number_of_samples
    end

    # initialize categorical distribution w/p -
    p = softmax(őł);
    data_distribution = Categorical(p);

    data_distribution,p    
end;
```

## Task 2: Feedforward and Feedback Passes
In this task, let's assume some random values for the parameters $\mathbf{W}$, $\mathbf{a}$, and $\mathbf{b}$ and see what happens when we do a combined _feedforward_ and _feedback_ pass through the restricted Boltzmann machine. First, let's build a model. We'll create an instance of [the `MyRestrictedBoltzmannMachineModel` type](https://varnerlab.github.io/VLDataScienceMachineLearningPackage.jl/dev/boltzmann/#VLDataScienceMachineLearningPackage.MyRestrictedBoltzmannMachineModel) with random values for the $\mathbf{W}$, $\mathbf{a}$, and $\mathbf{b}$ parameters.

> The [`randn(...)` method](https://docs.julialang.org/en/v1/stdlib/Random/#Base.randn) generates normally distributed random numbers with mean `0` and standard deviation `1`. 

We'll save the model in the `model::MyRestrictedBoltzmannMachineModel` variable.

```julia
model = let 
    
    # initialize some random weights and biases
    W = 2*randn(number_of_visible_nodes, number_of_hidden_nodes);
    W[diagind(W)] .= 0.0; # set the diagonal to zero
    a = randn(number_of_visible_nodes)
    b = randn(number_of_hidden_nodes);

    # next, build the model -
    model = build(MyRestrictedBoltzmannMachineModel, (
        W = W,
        a = a,
        b = b,
    ));

    # return -
    model;
end;
```

__Sample__. We've implemented the restricted Boltzmann machine's feedforward and feedback passes in [the `sample(...)` method](https://varnerlab.github.io/VLDataScienceMachineLearningPackage.jl/dev/boltzmann/#VLDataScienceMachineLearningPackage.sample-Tuple{MyRestrictedBoltzmannMachineModel,%20Vector{Int64}}).

> __What is going on?__ Break up into groups and describe to each other what is going on in the feedforward and feedback passes of the RBM. Check out [the `sample(...)` method documentation and code](https://varnerlab.github.io/VLDataScienceMachineLearningPackage.jl/dev/boltzmann/) for implementation details, or [the notes on block Gibbs sampling and contrastive divergence](labs/week-7/L7d/CHEME-5820-L7d-Advanced-Sampling-Training-RBM-Spring-2026.ipynb) for a more theoretical description.

The `sample(...)` method returns a tuple `(v, h)` containing the visible layer states $\mathbf{v}$ and hidden layer states $\mathbf{h}$ recorded at each of the `T` internal steps.

```julia
(v‚āí,v, h) = let

    # draw a random index from the data distribution
    i = rand(data_distribution);
    
    v‚āí = X[:,i]; # initial visible layer values
    (v,h) = VLDataScienceMachineLearningPackage.sample(model, v‚āí, T = number_of_internal_steps, ő≤ = ő≤);

    (v‚āí,v,h)
end;
```

From that sequence, we grab the last value, and then we generate $\mathbf{v}_{1},\dots,\mathbf{v}_{n}$ feedback steps _starting_ from $\mathbf{h}_{n}$.

## Task 3: Training with Contrastive Divergence
In this task, let's do a few training steps and see how this changes the behavior of the restricted Boltzmann machine. 

> __Contrastive Divergence (CD)__
>
> The training method for the RBM is known as contrastive divergence. The CD algorithm approximates the gradient of the log-likelihood by running a limited number of Gibbs sampling steps rather than waiting for equilibrium. See this [description of the CD algorithm]() for more details. We've implemented the CD algorithm in [the `learn(...)` method](https://varnerlab.github.io/VLDataScienceMachineLearningPackage.jl/dev/boltzmann/#VLDataScienceMachineLearningPackage.learn-Tuple{MyRestrictedBoltzmannMachineModel,%20Matrix{Int64},%20Categorical{P}%20where%20P%3C:Real}). 
> * The [`learn(...)` method](https://varnerlab.github.io/VLDataScienceMachineLearningPackage.jl/dev/boltzmann/#VLDataScienceMachineLearningPackage.learn-Tuple{MyRestrictedBoltzmannMachineModel,%20Matrix{Int64},%20Categorical{P}%20where%20P%3C:Real}) takes a `model::MyRestrictedBoltzmannMachineModel` instance, the image dataset `X::Array{Int,2}` array, our approximate value for $P(\mathbf{x})_{\text{data}}$, the system (inverse) temperature `ő≤::Float64`, the learning rate `ő∑::Float64` and some other control parameters, e.g., `maxnumberofiterations::Int` which is the maximum number of iterations we want the training to go for.

This method returns a new `MyRestrictedBoltzmannMachineModel` with the updated parameters.

```julia
newmodel = let

    # initialize -
    trained_model_filepath = joinpath(_PATH_TO_DATA, "trained_rbm_model.jld2");
    do_I_want_to_train_the_model = false; # Set to false to load pretrained model
    newmodel = nothing;
    localmaxnumberofiterations = 100;
    localbatchsize = 50;
    verboseflag = true;

    if (do_I_want_to_train_the_model == true)
        newmodel = learn(model,X,data_distribution, maxnumberofiterations = localmaxnumberofiterations,
            T = number_of_internal_steps, ő≤ = ő≤, ő∑ = ő∑, verbose = verboseflag, 
                batchsize = localbatchsize);
        
        # Save the trained model and associated data
        jldsave(trained_model_filepath; 
            W = newmodel.W, 
            a = newmodel.a, 
            b = newmodel.b,
            number_of_visible_nodes = number_of_visible_nodes,
            number_of_hidden_nodes = number_of_hidden_nodes
        );
        println("Model saved to: $(trained_model_filepath)")
    else
        # Load the pretrained model
        if isfile(trained_model_filepath)
            saved_data = load(trained_model_filepath);
            newmodel = build(MyRestrictedBoltzmannMachineModel, (
                W = saved_data["W"],
                a = saved_data["a"],
                b = saved_data["b"],
            ));
            println("Model loaded from: $(trained_model_filepath)")
        else
            error("Pretrained model file not found at $(trained_model_filepath). Please train the model first.")
        end
    end

    # return -
    newmodel;
end;
```

__After a single pass of the trained model__? Suppose we give $\bar{\mathbf{v}}_{\circ}$ to the model, and let it do `number_of_internal_steps::Int` number of internal steps, i.e., $n$-feedforward steps which generated $\bar{\mathbf{h}}_{1},\dots,\bar{\mathbf{h}}_{n}$, and then we generated $\bar{\mathbf{v}}_{1},\dots,\bar{\mathbf{v}}_{n}$ feedback steps _starting_ from $\bar{\mathbf{h}}_{n}$.

What do we get?

```julia
v‚āí, vŐĄ‚āí, vŐĄ, hŐĄ = let

    # draw a random index from the data distribution
    i = rand(data_distribution); # get a random index from the data distribution
    v‚āí = copy(X[:,i]); # initial visible layer values
    vŐĄ‚āí = copy(v‚āí); 
    local_number_of_internal_steps = 5; # we are going to sample back and fourth this many times

    # let's make this a little harder -
    for j ‚ąą eachindex(v‚āí)
        if (rand() ‚Č§ 0.6) # add a little noise to our image
            vŐĄ‚āí[j] = -1;
        end
    end

    # run the simulation (block Gibbs sampling)
    (v,h) = VLDataScienceMachineLearningPackage.sample(newmodel, v‚āí, 
        T = local_number_of_internal_steps, ő≤ = 0.1*ő≤);

    # return -
    v‚āí, vŐĄ‚āí, v, h
end;
```

## Summary
This lab explored the sampling and training of a restricted Boltzmann machine on handwritten digit images using contrastive divergence.

> __Key Takeaways__
>
> * **Feedforward and feedback passes:** An RBM generates hidden states from visible inputs (feedforward) and reconstructs visible states from hidden activations (feedback) through block Gibbs sampling. Each layer's states are sampled using conditional probability distributions computed from the opposite layer's activations.
> * **Contrastive divergence training:** The CD algorithm updates weights and biases by comparing correlations between data-driven activations and model reconstructions, avoiding the need to compute the intractable partition function. In this lab, the CD algorithm learns parameters that capture patterns in the USPS handwritten digit images through iterative training steps.
> * **Reconstruction improvement:** After training, the RBM produces reconstructions that better match the input data distribution, demonstrating that the network has learned the underlying structure of the images. This improvement is visible when comparing reconstructions of noisy or corrupted digits, where the trained model recovers cleaner versions of the original patterns.

The trained RBM can be used for tasks such as dimensionality reduction, feature learning, and generative modeling.
